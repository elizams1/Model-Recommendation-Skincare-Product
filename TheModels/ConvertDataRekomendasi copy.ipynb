```python
import pandas as pd
import numpy as np
```

```python
data_index = pd.read_excel('C:/Users/HP/Downloads/TUGAS AKHIR/EXCEL KODINGAN/IndexRekomendasi.xlsx')
data_index
```

```python
data_index.info()
```

```python
import ast
ast.literal_eval(data_index['index'][0])
```

```python
index = ast.literal_eval(data_index['index'][0])
# index = [10, 11, 17, 6, 3]
temp = []
for k in range(0, 100):
  if(k in index):
    res = 1
    temp.append(res)
  else:
    res = 0
    temp.append(res)

print(temp)
```

```python
res_index_sistem = pd.DataFrame({'index' : []})
for i in range(0,len(data_index)):
  temp = []
  az = ast.literal_eval(data_index['index'][i])
  for k in range(0,100):
      if(k in az):
        res = 1
        temp.append(res)
      else:
        res = 0
        temp.append(res)
  new_res = pd.DataFrame([{'index': temp}])
  res_index_sistem = pd.concat([res_index_sistem, new_res], ignore_index=True)
```
